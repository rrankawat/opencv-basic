###### Loading Image

### In this project, weâ€™ll learn:

* How to read and display images using OpenCV
* How to apply filters (blur, edge detection, thresholding, etc.)
* How to visualize them using matplotlib

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Helper function
def show(img, cmap=None):
  if len(img.shape) == 3:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
  plt.imshow(img, cmap)
  plt.axis('off')
  plt.show()
```

```python
# Read the image from your local folder
img = cv2.imread('./media/dog.jpg')
```

```python
# Check if image loaded correctly
if img is None:
    print('Error: Image not found. Check the path...')
    exit()
```

```python
# Display original image using Matplotlib
show(img)
```

###### Apply Filters

```python
# 1. Convert to Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show(gray, cmap='gray')
```

```python
# 2. Gaussion Blur
blur = cv2.GaussianBlur(img, (7, 7), 0)
show(blur)
```

```python
# 3. Edge Detection (Canny)
edges = cv2.Canny(gray, 100, 200)
show(edges, cmap='gray')
```

```python
# 4. Sharpening Filter
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
sharpened = cv2.filter2D(img, -1, kernel)
show(sharpened)
```

```python
# 5. Thresholding
thresh = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)[1]
show(thresh, cmap='gray')
```
